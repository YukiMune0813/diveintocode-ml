問題】栗まんじゅうが太陽系を覆う日
1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。
栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。
ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも適用できるような汎用的なプログラムを作成してください。また、グラフを表示するなど分かりやすいノートブックを作成してください。
宇宙空間は半径138億光年の球
体積80㎤の栗まんじゅう
体積=4/3*π*r**3
5分ごとに倍

```python
#宇宙の体積
import math
cosmos_r=1.30558080522e+28 #宇宙の半径cm
cosmos_v=4/3*math.pi*cosmos_r**3
print(cosmos_v)
```

```python
#栗まんじゅうが覆う汎用性あり
def bibine_v_versatility(object_volume ,target_volume):
    time_count = 0
    volume = object_volume
    volume_history = [volume]
    while target_volume > volume:
        volume *=2
        time_count += 5
        volume_history.append(volume)
        
    import matplotlib.pyplot as plt
    %matplotlib inline

    plt.title("Completely cover")
    plt.xlabel("time")
    plt.ylabel("volume[cm^3]")
    plt.plot(volume_history)
    plt.show()
        
    return time_count

```

```python
#ラグビーボールの体積を5087[cm^3]
print("ラグビーボールが宇宙空間の体積を埋め尽くす時間 : {}分" .format(bibine_v_versatility(5087,cosmos_v)))




```
