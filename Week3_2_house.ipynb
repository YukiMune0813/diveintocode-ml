```python
import missingno as msno

# 全体像を見る
msno.matrix(df)
```

### 欠損値が含有割合

```python
#特徴量ごとの欠損していない割合をみる
msno.bar(df)
```

```python
#欠損値を含む特徴量だけを表示。０以下の欠損値は表示しない
therehold=0 #この値以下の欠損値の特徴量は表示しない
n_samples = len(df)
missing_rate =df.isnull().sum() / n_samples*100
print("欠損値が{}%より高い特徴量の一覧\n{}".format(therehold, missing_rate[missing_rate>therehold]))
```

```python
import seaborn as sns
plt.figure(figsize=(16,20)) #サイズ調整

sns.heatmap(df.isnull(), cbar=False)
```

### ５つ以上欠損値のある特徴量(列)を削除

```python
therehold=0 #この値以下の欠損値の特徴量は表示しない
n_missing=df.isnull().sum()
print("欠損数が{}より高い特徴量の一覧\n{}".format(therehold, n_missing[n_missing>therehold]))
```

```python
#欠損数と欠損率を表にする
df_missing=pd.concat([n_missing, missing_rate], axis=1)#欠損数と欠損率をpd.concatで結合、axis=1は横に結合
df_missing=df_missing.rename(columns={0:"total", 1: "missing_ratio"})#カラムをリネーム、0列目"total"
df_missing=df_missing.query("total>0 and missing_ratio>0")#total,ratioが0より大きいもの
display(df_missing)

#df.rename(columns={'A': 'a'}, index={'ONE': 'one'})
```

```python
#欠損値が５つ以上ある特徴量を抽出します。
therehold=5
many_missing_features = df_missing.query("total >={}".format(therehold)).index#index表示
print("欠損値が{}個以上ある特徴量の一覧 :\n{}".format(therehold,many_missing_features ))
```

```python
#欠損が５個以上の列を削除、削除には.drop(第一引数labels,第二引数axisで指定)を使用する

df_dropped_features = df.drop(many_missing_features, axis=1)
display(df_dropped_features.head())
```

### 欠損値があるサンプル(行)を削除

```python
#　欠損の多い特徴量を消した上で、さらに欠損がある行を削除、行の削除はdropna()
df_dropna=df_dropped_features.dropna(axis=0)#行に対して欠損値を削除
df_dropna=df_dropna.reset_index(drop=True)
print("残っているサンプル数 : {}".format(len(df_dropna)))
msno.bar(df_dropna)
```

## 【問題5】用語の調査
目的変数がどのような分布を持っているか確認  
尖度：確率変数の確率密度関数や頻度分布の鋭さを表す。正規分布と比較して、尖度が大きければ鋭いピークと長く太い裾を持った分布をもち、尖度が小さければより丸みがかかったピークと短く細い尾を持った分布であることが判断できる。 
正規分布だと尖度3になる

歪度:分布の歪み度合いを示すもの。正規分布のように左右対称であるのか、それとも片方の裾だけ伸びたような歪んだ分布なのか、その度合いを示す。歪度が０よりも大きければ右の裾が長い分布、0よりも小さければ左の裾が長い分布


### 【問題6】分布の確認
- seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
- 目的変数に対して対数変換を行う。
- 対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。  
対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。
seaborn.distplot — seaborn 0.9.0 documentation

```python
#seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
target=df["SalePrice"]
print("歪度={:.2f} 尖度={:.2f}".format(target.skew(), target.kurtosis()))
display(sns.distplot(target, kde=False, rug=False, bins=10))#rugとは?
#kdeは密度近似関数の描画のオプション
```

```python
#目的変数に対して対数変換を行う。
target_log=target.apply(np.log) #.apply(np.log)?
#対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

print("歪度={:.2f} 尖度={:.2f}".format(target_log.skew(), target_log.kurtosis()))
display(sns.distplot(target_log, kde=False, rug=False, bins=10))
```

- 対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。
正規分布の場合、尖度 ＝ 3 となるので、これが基準である。元のデータより、対数変換したデータの方が基準の3に近い。対数変換すると正規分布に近づく場合が多いので、対数変換することが一般的である。

### 【問題7】相関係数の確認

特徴量の確認をしていきます。それぞれに対してマークダウンで説明や考察を書いてください。

- 相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。
- 目的変数との相関の高い特徴量10個を選び、これらだけの相関係数行列ヒートマップを作る。
- 選んだ10個の特徴量が何か表しているかをKaggleのDataDescriptionの記述を参考に日本語でまとめる。
- 選んだ10個の特徴量の中でお互いの相関係数が高い組み合わせを3つ探し出す。

**補足**

重回帰分析などで予測する場合、相関が高すぎる特徴量の組み合わせがあると予測に悪影響が出ることがあります。このような場合の簡単な対処方法として、そういった特徴量の組みを探し出して、１つに絞るということがあります。

```python
#相関係数行列のヒートマップを作り、相関係数の高い関係を見つける。
#ヒートマップを作るにはまずは相関行列関数.corr()で作成

df_corr=df.corr()
#df_corr

colormap=plt.cm.RdBu#赤と青
plt.subplots(figsize=(15,12))#わからない
plt.title("Correlation matrix")
sns.heatmap(df_corr, linewidths=0.1, vmax=1.0, square=True, cmap=colormap, linecolor="white", annot=False)#相関の最大値は1だからvmax=1.0
#annot=True セルに値を出力,行列などのデータセットを指定すると、任意の値を出力できます。(デフォルト値: None)
plt.show()
```

### 目的関数との相関関数の高い特徴量を１０個選び、これらだけでヒートマップを作る

```python
#目的関数、SalePriceと相関の高い特徴量の選択
df_corr_saleprice=df_corr["SalePrice"].sort_values(ascending=False)
#要素の値に応じてソートするにはsort_values()デフォルトは昇順。降順にしたい場合は引数ascendingをFalse
#print("目的関数と相関が高い順にソート\n{}".format(df_corr_saleprice))
print("目的関数との相関の高い特徴量１０個\n{}".format(df_corr_saleprice[1:11]))
      #スライス:0行目（saleprice同士）と11番目以降を削除
```
