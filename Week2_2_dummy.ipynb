# 【問題1】乱数の作成
平均が(-3, 0)、共分散行列が以下で表される2次元正規分布による乱数を500個作成してください。
[
1.0
0.8
0.8
1.0

]
 
多次元の正規分布による乱数はnp.random.multivariate_normal()を使うことで作成できます。
numpy.random.multivariate_normal — NumPy v1.16 Manual
作成したndarrayのshapeは(500, 2)になります。

?《シード値について》
実行ごとに値が変わらないように、np.random.seed()により乱数のシード値を固定しておきましょう。今回シード値は0を使うことにします。
np.random.seed(0)

```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
 
mean_data= [-3, 0]#平均
cov= [[1.0, 0.8], [0.8, 1.0]]#共分散行列
n_samples=500 #サンプル数
np.random.seed(0)#乱数のシード値を固定

# 2次元正規乱数を1万個生成
data1= np.random.multivariate_normal(mean_data, cov, n_samples)
# numpy.random.multivariate_normal(mean, cov, size, check_vali)
# loc: 生成したい正規分布の平均値ベクトルを指定
# scale: 生成したい正規分布の分散共分散行列を指定
# size: 生成する乱数の数を指定(default=1)
# check_valid: 分散共分散行列の半正定値チェック有無(warn/raise/ignoreを指定。オプション)
# tol: 半正定値チェックの許容誤差(オプション)

print("作成したデータ\n{}".format(data1))
```

【問題3】ヒストグラムによる可視化
問題1で作成したデータをヒストグラムにより可視化してください。
ヒストグラムはplt.hist()を使うことで描けます。
matplotlib.pyplot.hist — Matplotlib 3.1.1 documentation
各次元ごとに作成し、2つのヒストグラムとしてください。
それぞれのヒストグラムの横軸の範囲はplt.xlim()を用いて揃えるようにしてください。
matplotlib.pyplot.xlim — Matplotlib 3.1.1 documentation

ヒストグラムで正規分布であることがわかりやすくなる。
plt.xlim()で平均の違いがはっきりする

https://pythondatascience.plavox.info/matplotlib/ヒストグラム

```python
import numpy as np
import matplotlib.pyplot as plt

#ヒストグラムは特定のデータを区間ごとに区切り、各区間の個数やばらつきを棒グラフで表す。

#最小値.min()と最大値.max()を出して、それらを範囲に設定する。
data1_min=data1.min()
data1_max=data1.max()

plt.title("histogram of x1")
plt.xlabel("x1")
plt.ylabel("frequency")
plt.hist(data1[:,0],bins=50)
plt.xlim([data1_min, data1_max])#pyplot.xlim: x 軸方向の表示範囲を取得または指定
plt.show()

plt.title("histogram of x2")
plt.xlabel("x12")
plt.ylabel("frequency")
plt.hist(data1[:,1],bins=50)
plt.xlim([data1_min, data1_max])#pyplot.xlim: x 軸方向の表示範囲を取得または指定
plt.show()
```

【問題5】データの結合
データはまとめておいた方が後々扱いやすいです。問題1、4で作成したndarrayを 結合 し、(1000, 2)のndarrayとしてください。結合はnp.concatenate()やnp.vstack()を使うことで行えます。
numpy.concatenate — NumPy v1.16 Manual
numpy.vsplit — NumPy v1.16 Manual

#ndarrayの結合　https://note.nkmk.me/python-numpy-concatenate-stack-block/

・既存の軸（次元）に沿って結合するのがnumpy.concatenate()、2次元配列を縦横に結合
第一引数に結合したいndarrayのリストを指定、第二引数axisに結合する軸（次元）
デフォルトはaxis=0なので行、つまり縦に結合
axis=1とすると2番目の軸（1次元目）が列なので横に結合

・新たな軸に沿って結合するのがnumpy.stack()、2次元配列を重ねて3次元配列を生成するのがnumpy.stack()

・np.vstack()は縦に重ねる

・np.hstack()を使って横方向に配列を結合

```python
data=np.concatenate([data1,data2], axis=0)
print(data)
```

```python
data=np.vstack([data1,data2])
print(data)
```

【問題6】ラベル付け
ひとまとめになった1000個のデータそれぞれに対して、問題1、4どちらで作成したものなのかを示す ラベル 付けを行ってください。問題1のものには0、問題4のものには1を対応させます。
0と1を含むラベルの列を新たに追加し、(1000, 3)のndarrayを作成してください。
機械学習に使用するデータセットはこのような形になっていることが多いです。

np.zeros()で0、np.ones()を使い1のndarrayを作る方法
numpy.zeros(shape(x,y), dtype=float, order='C')

NumPyのデータ型dtype一覧とastypeによる変換（キャスト）
https://note.nkmk.me/python-numpy-dtype-astype/

```python
#np.zeros()で0の、np.ones()で1のndarray(N次元配列を扱うためのクラス)を作成
#0が500行、1が500行のndarrayをnp.vstackで結合
label_data1=np.zeros((n_samples,1))
label_data2=np.ones((n_samples,1))

#label_data1=np.fill((n_samples,1),0) np.full((shape), 入れる数字)

label=np.vstack([label_data1,label_data2])

#labelをdataの３列目に入れる、np.concatenate([,],axis=1)
data_with_label=np.concatenate([data, label], axis=1)

print("label化したndarray\n{}".format(data_with_label))

#data_with_label=np.concatenate([data, label[:,np.newaxis]], axis=1)、np.newaxisで軸を増やす

```

np.empty((shape))で空のndarrayを作成しスライスで0と1を埋めていく
https://deepage.net/features/numpy-empty.html

```python
label=np.empty((data.shape[0],1)).astype(np.int)
label[:n_samples]=0
label[n_samples:]=1
```
