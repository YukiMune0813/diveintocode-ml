【問題1】100日目の米粒の数
100日目には合計で米粒をいくつもらっているのかを計算するコードを作成し、答えを出力してください。また、1日目から100日目までの「その日にもらう米の数」と「その日までに累計で貰う米の数」の推移を折れ線グラフで表現してください。（縦軸が米の数、横軸が日数）



```python
day_rice=0
day=100
rice_list=[]
rice_sum_list=[]
#rice_count.list=[]
for i in range(day-1):
    rice_list.append(1*2**day_rice)
    day_rice+=1
    rice_sum_list.append(sum(rice_list))
print("100日目の米粒の数は{}粒です。".format(sum(rice_list)))
```

```python
##その日までに累計で貰えるコメ のグラフ化
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("numbers of rice")
plt.xlabel("day")
plt.ylabel("rice")
plt.plot(rice_sum_list) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```
